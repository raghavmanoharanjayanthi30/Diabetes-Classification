```python
from sklearn.model_selection import RandomizedSearchCV
est = RandomForestClassifier()
parameters = {"max_depth": [1,2,3,4,5,6,8,10,None],
                 "n_estimators": [50,100,125,135,150,160,175,200,250,300,400],
                 "max_features":[1,2,3,4],
                 "criterion": ["gini", "entropy","log_loss"],
                 "bootstrap": [True, False],
                 "min_samples_leaf":[1,2,3,4,5,6,7,8]
                }
def hypertuning_rscv(est, p_dist, numIter, X, y):
    rdmSearch = RandomizedSearchCV(est, param_distributions= p_dist, n_jobs=-1, n_iter=numIter,cv = 9)
    
    rdmSearch.fit(X,y)
    ht_params = rdmSearch.best_params_
    ht_score = rdmSearch.best_score_
    return ht_params, ht_score


```

```python
rf_parameters, rf_ht_score = hypertuning_rscv(est,parameters,40,X_train,Y_train)
```
